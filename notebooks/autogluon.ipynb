```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
fold = 0
```

```python
# Load the data
df = pd.read_csv("../input/train_folds.csv")

# Split the data into training and testing sets
train = df[df.kfold != fold].reset_index(drop=True)
test = df[df.kfold == fold].reset_index(drop=True)

# Split the data into X and y
train = train.drop(["id", "kfold"], axis=1)
test = test.drop(["id", "kfold"], axis=1)
```

```python
print("Train columns:", train.columns)
print("Test columns:", test.columns)
```

```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
predictor = TabularPredictor(label="Response", verbosity=3, eval_metric='roc_auc').fit(
    train_data=train,
    time_limit=1800,
    presets="optimize_for_deployment",
    ag_args_fit={'max_memory': 30}
    )
```

```python
predictor.leaderboard()
```

```python
# Report metrics on the test set
data = test.drop("Response", axis=1)

# Evaluate metrics
extra_metrics = [roc_auc_score]
```

```python
# Prediciting
test_data = TabularDataset(data)
y_pred = predictor.predict(test_data)
y_true = test["Response"]
print(roc_auc_score(y_true, y_pred))
```

```python
y_pred_lgbm = predictor.predict(test_data, model="LightGBM")
print(roc_auc_score(y_true, y_pred_lgbm))
```

```python
models_to_compare = ['LightGBM', 'LightGBMXT', 'KNeighborsUnif']  # Add the top models from the leaderboard
for model_name in models_to_compare:
    y_pred_model = predictor.predict(test_data, model=model_name)
    score = roc_auc_score(y_true, y_pred_model)
    print(f"Model: {model_name}, ROC-AUC: {score}")
```
